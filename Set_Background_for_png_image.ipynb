```python
import cv2 as cv
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import pylab
pylab.rcParams['figure.figsize']=(7,7)
```

```python
def set_background_for_png_image(image, file_name):
    image = cv.imread(image, cv.IMREAD_UNCHANGED) #Use IMREAD_UNCHANGED to ensure loading of alpha channel
    trans_mask = image[:,:,3] == 0
    image[trans_mask] = [255, 255, 255, 255]

    # Save Image
    cv.imwrite(str(file_name)+".png", image)

    # Visualize
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    print(image.shape)
    plt.axis("off")
    plt.show()

set_background_for_png_image("/content/Fernanda.png", "anak")
```
